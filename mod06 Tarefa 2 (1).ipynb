####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Caminho do arquivo SINASC.csv
caminho_arquivo = r'C:\Users\Acer\Desktop\pratique 2\SINASC.csv'

# Carregar os dados
dados = pd.read_csv(caminho_arquivo)

# Plotar o boxplot
plt.figure(figsize=(8,6))
plt.boxplot(dados['IDADEMAE'])
plt.title('Boxplot da Idade da Mãe')
plt.ylabel('Idade')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho do arquivo SINASC.csv
caminho_arquivo = r'C:\Users\Acer\Desktop\pratique 2\SINASC.csv'

# Carregar os dados
dados = pd.read_csv(caminho_arquivo)

# Plotar o histograma da idade da mãe
plt.figure(figsize=(8,6))
plt.hist(dados['IDADEMAE'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```
